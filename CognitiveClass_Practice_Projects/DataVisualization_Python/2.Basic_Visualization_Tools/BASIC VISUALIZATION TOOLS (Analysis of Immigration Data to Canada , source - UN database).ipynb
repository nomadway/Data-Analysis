# Display Top 3 countries that sent most immigrants to Canada

```python
canad3 = canad.head(3)
```

```python
canad3
```

# ***BASIC VISUALIZATION TOOLS***

#                                          Area Plot

```python
##Option-1
canad3.index = canad3.index.map(int)

canad3.plot(kind='area', alpha=0.5, stacked=False, figsize=(12,6))

plt.title('Immigration From Top 3 Countries to Canada (1993-2013)')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')


plt.show()
```

```python
# Option 2: This a preferred option as it provides more flexibility
##Option 2: Artist layer (Object oriented method) - using an Axes instance from Matplotlib (preferred) **
##You can use an Axes instance of your current plot and store it in a variable (eg. ax). 
##You can add more elements by calling methods with a little change in syntax 
#(by adding "set_" to the previous methods). 
#For example, use ax.set_title() instead of plt.title() to add title, or ax.set_xlabel() instead of plt.xlabel() to add label to the x-axis.
ax = canad3.plot(kind='area', alpha=0.35, figsize=(12, 6))

ax.set_title('Immigration From Top 3 Countries to Canada (1993-2013)')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

# Histograms
(is a way of representing the frequency distribution of a variable)

```python
canad['2013'].head()
```

```python
count, bin_edges = np.histogram(canad['2013'])

print(count) 
print(bin_edges) 



#     178 countries contributed between 0 to 3412.9 immigrants
#     11 countries contributed between 3412.9 to 6825.8 immigrants
#     1 country contributed between 6285.8 to 10238.7 immigrants,etc.
```

```python
count, bin_edges = np.histogram(canad['2013'])

canad['2013'].plot(kind='hist', figsize=(12, 6), xticks=bin_edges)

plt.title('Immigration from 195 countries to Canada in 2013') 
plt.ylabel('Number of Countries') 
plt.xlabel('Number of Immigrants')

plt.show()
```

# Show immigration distribution for Russia, Ukraine, and Belarus for years 1993-2013.

```python
canad.loc[['Russian Federation', 'Ukraine', 'Belarus'], years]
```

```python
canad.loc[['Russian Federation', 'Ukraine', 'Belarus'], years].plot.hist()
#Need to transpose() the dataframe
```

```python
canad.loc[['Russian Federation', 'Ukraine', 'Belarus'], years].transpose()
ru_uk_bel = canad.loc[['Russian Federation', 'Ukraine', 'Belarus'], years].transpose()
ru_uk_bel.head()
```

```python
count, bin_edges = np.histogram(ru_uk_bel, 15)
xmin = bin_edges[0] - 10 
xmax = bin_edges[-1] + 10


ru_uk_bel.plot(kind ='hist', 
          figsize=(12, 6),
          bins=15,
          alpha=0.6,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen'],
              stacked=True,
                xlim=(xmin, xmax)
              )

plt.title('Immigration from Russia, Ukraine, and Belarus to Canada (1993 - 2013)')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

# Immigration from Top 15 Countries to Canada

```python
years = list(map(str,range(1993, 2014)))
canad.sort_values(['Total'], ascending=False, axis=0, inplace=True)
canad15 = canad['Total'].head(15)
canad15
```

```python
canad15.plot(kind='barh', figsize=(12, 6), color='steelblue')

plt.xlabel('Number of Immigrants')
plt.ylabel('Countries')
plt.title('Immigration From Top 15 Countries to Canada (1993-2013)')

   # annotate value labels to each country
for index, value in enumerate(canad15): 
    label = format(int(value), ',') # format int with commas
      # place text at the end of bar (subtracting 47000 from x, and 0.1 from y to make it fit within the bar)
    plt.annotate(label, xy=(value - 47000, index - 0.10), color='white')


plt.show()
```
