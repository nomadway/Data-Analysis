# Compare the number of immigrants from Russian and Ukraine from 1993 to 2013.

```python
canad.loc['Russian Federation', years]
```

```python
canad.loc['Ukraine', years]
```

```python
canad.loc[['Russian Federation', 'Ukraine'], years] ##Need to assign a variable

ruuk = canad.loc[['Russian Federation', 'Ukraine'], years] 
ruuk
```

```python
ruuk.plot(kind='line')
```

```python
#The country is the index and years is column in the dataframe. Thus need to use the transpose() method to 
#swap the rows and columns 
ruuk = ruuk.transpose()
ruuk.head()
```

```python
ruuk.index = ruuk.index.map(int) # change the index values of ruuk to type integer for plotting
ruuk.plot(kind='line', figsize=(10,5))

plt.title('Immigration from Russia and Ukraine to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
