```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis = 'columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country': 'TotalRespondents', 'LanguageWorkedWith': 'KnowsPython'}, inplace=True)
python_df
```

```python
python_df['PercentKnowsPython'] = (python_df['KnowsPython'] / python_df['TotalRespondents']) * 100
python_df
```

```python
python_df.sort_values(by='PercentKnowsPython', ascending=False, inplace=True) #sort countries that know python more 
python_df.head(20)
```
