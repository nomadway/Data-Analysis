# 20. Classify Movies Based on Ratings such as 'Excellent', 'Good', and 'Average'

```python
movie.columns
```

```python
def rating (rating):
    if rating >= 8.0:
        return 'Excellent'
    elif rating >= 6.0:
        return 'Good'
    else:
        return 'Average'
```

```python
##Use 'apply' method and the 'rating' from function, and assign a new column variable 'movie_category'
movie['Rating_Category'] = movie['Rating'].apply(rating)
```

# 22. Display Unique Values From 'Genre' Column

```python
movie['Genre']
#This task to be solved in THREE steps
```

```python
#STEP-1 (Need to split unique values by a coma ',')
list = []
for value in movie ['Genre']:
    list.append(value.split(','))
```

```python
list
```

```python
##STEP-2 (sepeatin each value)
one_value = []
for item in list:
    for item1 in item:
        one_value.append(item1)
```

```python
one_value
```

```python
###STEP-3 (Find 'unique' values available in 'Genre' column)
```

```python
list2 = []
for item in one_value:
    if item not in list2:
        list2.append(item)
```

```python
list2
```

```python
len(list2)
```
